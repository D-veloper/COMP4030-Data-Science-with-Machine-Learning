```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
columns = ['Timestamp', 'x accelerometer', 'y accelerometer', 
           
           'z accelerometer', 'x gyro', 'y gyro', 'z gyro']
```

```python
grass_df = pd.read_csv('10_grass_moist_slopy.csv', header = None)

concrete_df = pd.read_csv('2_concrete_dry_bumpy_flat_hard.csv', header = None)

asphalt_df = pd.read_csv('5_asphalt_dry_flat_hard.csv', header = None)
```

```python
grass_df.columns = columns

concrete_df.columns = columns

asphalt_df.columns = columns
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

ax[0].plot(grass_df['x accelerometer'], color = 'green')

ax[0].set_title('X Accelerometer')

ax[0].set_xlabel('Sample Number')

ax[0].set_ylabel('X Accelerometer')


ax[1].plot(grass_df['y accelerometer'], color = 'green')

ax[1].set_title('Y Accelerometer')

ax[1].set_xlabel('Sample Number')

ax[1].set_ylabel('Y Accelerometer')


ax[2].plot(grass_df['z accelerometer'], color = 'green')

ax[2].set_title('Z Accelerometer')

ax[2].set_xlabel('Sample Number')

ax[2].set_ylabel('Z Accelerometer')


ax[3].plot(grass_df['x gyro'], color = 'green')

ax[3].set_title('X Gyro ')

ax[3].set_xlabel('Sample Number')

ax[3].set_ylabel('X Gyro')


ax[4].plot(grass_df['y gyro'], color = 'green')

ax[4].set_title('Y Gyro ')

ax[4].set_xlabel('Sample Number')

ax[4].set_ylabel('Y Gyro')


ax[5].plot(grass_df['z gyro'], color = 'green')

ax[5].set_title('Z Gyro ')

ax[5].set_xlabel('Sample Number')

ax[5].set_ylabel('Z Gyro')


plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

ax[0].plot(concrete_df['x accelerometer'], color = 'grey')

ax[0].set_title('X Accelerometer')

ax[0].set_xlabel('Sample Number')

ax[0].set_ylabel('X Accelerometer')


ax[1].plot(concrete_df['y accelerometer'], color = 'grey')

ax[1].set_title('Y Accelerometer')

ax[1].set_xlabel('Sample Number')

ax[1].set_ylabel('Y Accelerometer')


ax[2].plot(concrete_df['z accelerometer'], color = 'grey')

ax[2].set_title('Z Accelerometer')

ax[2].set_xlabel('Sample Number')

ax[2].set_ylabel('Z Accelerometer')


ax[3].plot(concrete_df['x gyro'], color = 'grey')

ax[3].set_title('X Gyro ')

ax[3].set_xlabel('Sample Number')

ax[3].set_ylabel('X Gyro')


ax[4].plot(concrete_df['y gyro'], color = 'grey')

ax[4].set_title('Y Gyro ')

ax[4].set_xlabel('Sample Number')

ax[4].set_ylabel('Y Gyro')


ax[5].plot(concrete_df['z gyro'], color = 'grey')

ax[5].set_title('Z Gyro ')

ax[5].set_xlabel('Sample Number')

ax[5].set_ylabel('Z Gyro')


plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

ax[0].plot(asphalt_df['x accelerometer'], color = 'black')

ax[0].set_title('X Accelerometer')

ax[0].set_xlabel('Sample Number')

ax[0].set_ylabel('X Accelerometer')


ax[1].plot(asphalt_df['y accelerometer'], color = 'black')

ax[1].set_title('Y Accelerometer')

ax[1].set_xlabel('Sample Number')

ax[1].set_ylabel('Y Accelerometer')


ax[2].plot(asphalt_df['z accelerometer'], color = 'black')

ax[2].set_title('Z Accelerometer')

ax[2].set_xlabel('Sample Number')

ax[2].set_ylabel('Z Accelerometer')


ax[3].plot(asphalt_df['x gyro'], color = 'black')

ax[3].set_title('X Gyro ')

ax[3].set_xlabel('Sample Number')

ax[3].set_ylabel('X Gyro')


ax[4].plot(asphalt_df['y gyro'], color = 'black')

ax[4].set_title('Y Gyro ')

ax[4].set_xlabel('Sample Number')

ax[4].set_ylabel('Y Gyro')


ax[5].plot(asphalt_df['z gyro'], color = 'black')

ax[5].set_title('Z Gyro ')

ax[5].set_xlabel('Sample Number')

ax[5].set_ylabel('Z Gyro')


plt.show()
```
