```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
columns = ['Timestamp', 'x accelerometer', 'y accelerometer', 
           
           'z accelerometer', 'x gyro', 'y gyro', 'z gyro']
```

```python
grass_df = pd.read_csv('../datasets/dayo/25_grass_slopy_wet.csv', header = None)

concrete_df = pd.read_csv('../datasets/dayo/13_concrete_dry_flat.csv', header = None)

asphalt_df = pd.read_csv('../datasets/dayo/19_asphalt_flat_wet.csv', header = None)

tile_df = pd.read_csv('../datasets/dayo/21_tile_smooth.csv', header = None)

stone_df = pd.read_csv('../datasets/dayo/16_stones_coarse.csv', header = None)
```

```python
grass_df.columns = columns

concrete_df.columns = columns

asphalt_df.columns = columns

tile_df.columns = columns

stone_df.columns = columns
```

```python
sample_grass_df = grass_df[1500:3500]

sample_grass_df.reset_index(drop = True, inplace = True)

sample_concrete_df = concrete_df[1450:3450]

sample_concrete_df.reset_index(drop = True, inplace = True)

sample_asphalt_df = asphalt_df[980:2980]

sample_asphalt_df.reset_index(drop = True, inplace = True)

sample_tile_df = tile_df[1760:3760]

sample_tile_df.reset_index(drop = True, inplace = True)

sample_stone_df = stone_df[1240:3240]

sample_stone_df.reset_index(drop = True, inplace = True)
```

```python
def y_lim(feature):
    
    if feature == 'x accelerometer':
        
        return (-1, 1)
        
    elif feature == 'y accelerometer':
        
        return (-0.8, 0.6)
        
    elif feature == 'z accelerometer':
        
        return (-2.25, 0)
        
    elif feature == 'x gyro':
        
        return (-30, 25)
        
    elif feature == 'y gyro':
        
        return (-15, 15)
        
    else:
        
        return (-15, 15)
```

```python
plt.figure(figsize = (20, 20))

for i, feature in enumerate(sample_grass_df.columns[1:], 1):
    
    plt.subplot(3, 2, i)
    
    sns.lineplot(sample_grass_df[feature], color = 'green')
    
    plt.ylim(y_lim(feature))
    
    plt.title(f'Distribution of {feature}')
```

```python
plt.figure(figsize = (20, 20))

for i, feature in enumerate(sample_concrete_df.columns[1:], 1):
    
    plt.subplot(3, 2, i)
    
    sns.lineplot(sample_concrete_df[feature], color = 'grey')
    
    plt.ylim(y_lim(feature))
    
    plt.title(f'Distribution of {feature}')
```

```python
plt.figure(figsize = (20, 20))

for i, feature in enumerate(sample_asphalt_df.columns[1:], 1):
    
    plt.subplot(3, 2, i)
    
    sns.lineplot(sample_asphalt_df[feature], color = 'black')
    
    plt.ylim(y_lim(feature))
    
    plt.title(f'Distribution of {feature}')
```

```python
plt.figure(figsize = (20, 20))

for i, feature in enumerate(sample_tile_df.columns[1:], 1):
    
    plt.subplot(3, 2, i)
    
    sns.lineplot(sample_tile_df[feature], color = 'wheat')
    
    plt.ylim(y_lim(feature))
    
    plt.title(f'Distribution of {feature}')
```

```python
plt.figure(figsize = (20, 20))

for i, feature in enumerate(sample_stone_df.columns[1:], 1):
    
    plt.subplot(3, 2, i)
    
    sns.lineplot(sample_stone_df[feature], color = 'saddlebrown')
    
    plt.ylim(y_lim(feature))
    
    plt.title(f'Distribution of {feature}')
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

for i, feature in enumerate(asphalt_df.columns[1:], 0):
    
    ax[i].plot(asphalt_df[feature], color = 'black')
    
    ax[i].set_title(feature)
    
    ax[i].set_xlabel('Sample Number')
    
    ax[i].set_ylabel(feature)

plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

features_to_plot = sample_grass_df.columns[1:]

for i, feature in enumerate(features_to_plot, 0):
    
    data_to_plot = pd.DataFrame({'Concrete': sample_concrete_df[feature], 'Grass': sample_grass_df[feature],
                                 
                                 'Asphalt': sample_asphalt_df[feature], 'Tile': sample_tile_df[feature],
                                
                                'Stone': sample_stone_df[feature]})
    
    sns.boxplot(data = data_to_plot, ax = ax[i])
    
    ax[i].set_xticks([0, 1, 2, 3, 4])
    
    ax[i].set_xticklabels(['Concrete', 'Grass', 'Asphalt', 'Tile', 'Stone'])
    
    ax[i].set_title(f'Box Plot of {feature}')
```
