```python
import pandas as pd
```

```python
columns = ['Timestamp', 'x accelerometer', 'y accelerometer', 
           
           'z accelerometer', 'x gyro', 'y gyro', 'z gyro', 'surface']
```

```python
### Grass

grass_23_df = pd.read_csv('../datasets/combined/23_GSW_trimmed.csv', header = None)

grass_23_df.columns = columns
```

```python
grass_24_df = pd.read_csv('../datasets/combined/24_GSW_trimmed.csv')

grass_24_df['surface'] = 'grass'

grass_24_df.columns = columns
```

```python
grass_25_df = pd.read_csv('../datasets/combined/25_GSW_trimmed.csv')
```

```python
grass_df = pd.concat([grass_23_df, grass_24_df, grass_25_df], axis = 0)
```

```python
grass_df
```

```python
grass_df['surface'].unique()
```

```python
grass_df.drop(['Timestamp'], axis = 1, inplace = True)

grass_df
```

```python
grass_df.to_csv('../datasets/final/GSW.csv', index = False)
```
