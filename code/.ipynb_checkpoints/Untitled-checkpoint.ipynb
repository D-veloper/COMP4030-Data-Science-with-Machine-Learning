```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame and it has columns named 'x_accelerometer', 'y_accelerometer', 'z_accelerometer'

df = pd.read_csv('../datasets/final/afw.csv') # Reading the asphalt dataset

# Define the sampling rate (fs)
fs = 100.0  # replace with your actual sampling rate

# Perform FFT on each accelerometer axis and store the positive frequencies and amplitudes
fft_results = {}
for axis in ['x accelerometer']:
    # Apply FFT to the signal
    fft_result = np.fft.fft(df[axis])
    fft_freq = np.fft.fftfreq(len(df[axis]), 1/fs)
    
    # Filter out the negative frequencies
    positive_mask = fft_freq > 0
    
    # Store the results in a dictionary
    fft_results[axis] = {
        'freq': fft_freq[positive_mask],
        'amplitude': np.abs(fft_result[positive_mask])
    }

# Now you can plot the results to visualize the frequency content
plt.figure(figsize=(15, 5))
for axis, result in fft_results.items():
    plt.plot(result['freq'], result['amplitude'], label=f'FFT {axis}')

plt.title('Frequency Domain Representation')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
