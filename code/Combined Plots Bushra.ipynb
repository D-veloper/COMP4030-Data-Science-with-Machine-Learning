```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
columns = ['Timestamp', 'x accelerometer', 'y accelerometer', 
           
           'z accelerometer', 'x gyro', 'y gyro', 'z gyro']
```

```python
grass_df = pd.read_csv('10_grass_moist_slopy.csv', header = None)

concrete_df = pd.read_csv('2_concrete_dry_bumpy_flat_hard.csv', header = None)

asphalt_df = pd.read_csv('5_asphalt_dry_flat_hard.csv', header = None)
```

```python
grass_df.columns = columns

concrete_df.columns = columns

asphalt_df.columns = columns
```

```python
plt.figure(figsize=(10, 6))

plt.plot(grass_df['x accelerometer'], label='x accelerometer')

plt.plot(grass_df['y accelerometer'], label='y accelerometer')

plt.plot(grass_df['z accelerometer'], label='z accelerometer')

plt.plot(grass_df['x gyro'], label='x gyro')

plt.plot(grass_df['y gyro'], label='y gyro')

plt.plot(grass_df['z gyro'], label='z gyro')

plt.legend()

plt.title('Grass Readings')

plt.xlabel('Sample Number')

plt.ylabel('Readings')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(concrete_df['x accelerometer'], label='x accelerometer')

plt.plot(concrete_df['y accelerometer'], label='y accelerometer')

plt.plot(concrete_df['z accelerometer'], label='z accelerometer')

plt.plot(concrete_df['x gyro'], label='x gyro')

plt.plot(concrete_df['y gyro'], label='y gyro')

plt.plot(concrete_df['z gyro'], label='z gyro')

plt.legend()

plt.title('Concrete Readings')

plt.xlabel('Sample Number')

plt.ylabel('Readings')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(asphalt_df['x accelerometer'], label='x accelerometer')

plt.plot(asphalt_df['y accelerometer'], label='y accelerometer')

plt.plot(asphalt_df['z accelerometer'], label='z accelerometer')

plt.plot(asphalt_df['x gyro'], label='x gyro')

plt.plot(asphalt_df['y gyro'], label='y gyro')

plt.plot(asphalt_df['z gyro'], label='z gyro')

plt.legend()

plt.title('Asphalt Readings')

plt.xlabel('Sample Number')

plt.ylabel('Readings')

plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

features_to_plot = grass_df.columns[1:]

for i, feature in enumerate(features_to_plot, 0):
    
    data_to_plot = pd.DataFrame({'Concrete': concrete_df[feature], 'Grass': grass_df[feature],
                                 
                                 'Asphalt': asphalt_df[feature]})
    
    sns.boxplot(data = data_to_plot, ax = ax[i])
    
    ax[i].set_xticks([0, 1, 2])
    
    ax[i].set_xticklabels(['Concrete', 'Grass', 'Asphalt'])
    
    ax[i].set_title(f'Box Plot of {feature}')
```
