```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
columns = ['Timestamp', 'x accelerometer', 'y accelerometer', 
           
           'z accelerometer', 'x gyro', 'y gyro', 'z gyro']
```

```python
grass_df = pd.read_csv('../datasets/praise/3_asphalt_dry_flat_hard.csv', header = None)

concrete_df = pd.read_csv('../datasets/praise/7_grass_moist_slopy.csv', header = None)

asphalt_df = pd.read_csv('../datasets/praise/9_concrete_dry_bumpy_flat_hard.csv', header = None)
```

```python
grass_df.columns = columns

concrete_df.columns = columns

asphalt_df.columns = columns
```

```python
for i, feature in enumerate(grass_df.columns[1: ], 0):
    
    print(i, feature)
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

for i, feature in enumerate(grass_df.columns[1:], 0):
    
    ax[i].plot(grass_df[feature], color = 'green')
    
    ax[i].set_title(feature)
    
    ax[i].set_xlabel('Sample Number')
    
    ax[i].set_ylabel(feature)

plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

for i, feature in enumerate(concrete_df.columns[1:], 0):
    
    ax[i].plot(concrete_df[feature].iloc[1050:2050], color = 'grey')
    
    ax[i].set_title(feature)
    
    ax[i].set_xlabel('Sample Number')
    
    ax[i].set_ylabel(feature)

plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

for i, feature in enumerate(asphalt_df.columns[1:], 0):
    
    ax[i].plot(asphalt_df[feature], color = 'black')
    
    ax[i].set_title(feature)
    
    ax[i].set_xlabel('Sample Number')
    
    ax[i].set_ylabel(feature)

plt.show()
```

```python
fig, ax = plt.subplots(6, 1, figsize=(13, 30), gridspec_kw={'hspace': 0.4})

features_to_plot = grass_df.columns[1:]

for i, feature in enumerate(features_to_plot, 0):
    
    data_to_plot = pd.DataFrame({'Concrete': concrete_df[feature], 'Grass': grass_df[feature],
                                 
                                 'Asphalt': asphalt_df[feature]})
    
    sns.boxplot(data = data_to_plot, ax = ax[i])
    
    ax[i].set_xticks([0, 1, 2])
    
    ax[i].set_xticklabels(['Concrete', 'Grass', 'Asphalt'])
    
    ax[i].set_title(f'Box Plot of {feature}')
```
